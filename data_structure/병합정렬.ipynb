- 잘게 쪼개고 합치면서 정렬하는 방법 
- 재귀 용법을 사용

```python
# split 과 merge

def merge_split(data):
    # data가 하나만 남을 때까지 
    if len(data) <= 1:
        return data
    # 중간을 기준으로 left, right로 나눔
    median = len(data) // 2
    # 재귀적으로 돌리고 -> 결국 하나만 남음
    left = merge_split(data[:median])
    right = merge_split(data[median:])
    
    return merge(left, right)
```

```python
def merge(left, right):
    
    merged_list = []
    l_pointer = 0
    r_pointer = 0
    
    # 왼쪽 오른쪽 둘 다 남아 있으면 돌려돌려 반복문
    while len(left) > l_pointer and len(right) > r_pointer:
        # 왼쪽 오른쪽 중에 작은 값을 먼저 넣어준다
        if left[l_pointer] > right[r_pointer]:
            merged_list.append(right[r_pointer])
            r_pointer += 1
        else:
            merged_list.append(left[l_pointer])
            l_pointer += 1
    
    while len(left) > l_pointer:
        # 왼쪽 ㅁ
        merged_list.append(left[l_pointer])
        l_pointer += 1
    
    while len(right) > r_pointer:
        merged_list.append(right[r_pointer])
        r_pointer += 1
    
    return merged_list
```

```python
def merge_sort(data):
    return merge_split(data)
```
