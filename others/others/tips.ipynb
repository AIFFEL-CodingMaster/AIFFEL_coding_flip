```python
%%writefile ../rcp.py

# import random 
from random import randint
n = int(input("몇 판 ?"))


# player_win, player_lose, computer_win, computer_lose, draw =0, 0, 0, 0, 0  


player_win, player_lose = 0, 0
computer_win, computer_lose = 0, 0
draw = 0


while n > 0:
    player = input("가위 바위 보 ?!?! : ")
    if player not in ['가위','바위','보']:
        continue
    computer = randint(0,2)


    if player == '가위':
        player = 0
    elif player == '바위':
        player = 1
    else:
        player = 2
    
    
    if player == computer:
        print("경기 결과 : 비김 ")
        draw += 1
    
    elif player == 0 and computer == 1:
        print("컴퓨터 : 바위")
        print("경기 결과 : 컴퓨터 승")
        player_lose += 1
        computer_win += 1

    elif player == 0 and computer == 2:
        print("컴퓨터 : 보")
        print("경기 결과 : 플레이어 승")
        player_win += 1
        computer_lose += 1

    elif player == 1 and computer == 0:
        print("컴퓨터 : 가위")
        print("경기 결과 : 플레이어 승")
        player_win += 1
        computer_lose += 1

    elif player == 1 and computer == 2:
        print("컴퓨터 : 보")
        print("경기 결과 : 컴퓨터 승")
        player_lose += 1
        computer_win += 1

    elif player == 2 and computer == 0:
        print("컴퓨터 : 가위")
        print("경기 결과 : 컴퓨터 승")
        player_lose += 1
        computer_win += 1

    elif player == 2 and computer == 1:
        print("컴퓨터 : 바위")
        print("경기 결과 : 플레이어 승")
        player_win += 1
        computer_lose += 1

    print()
    n -=1      
print(f"플레이어 {player_win}승 {draw}무 {player_lose}패")
print(f"컴퓨터   {computer_win}승 {draw}무 {computer_lose}패")
```
